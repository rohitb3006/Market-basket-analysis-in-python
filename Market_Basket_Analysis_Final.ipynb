```python
import pandas as pd #Used for importing the dataframe
import numpy as np #Used to perform operations in an numpy array form
from apyori import apriori #The library imported from the 
```

```python
dataset=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
dataset
```

```python
dataset.fillna(0,inplace=True)
dataset
```

```python
transac= []
for i in range(0,len(dataset)):
    transac.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
transac_rules = apriori(transac,min_support=0.005,min_confidance=0.3,min_lift=3,min_length=2)
```

```python
res=list(transac_rules)
```

```python
res
```

```python
dataset_results = pd.DataFrame(res)
```

```python
dataset_results
```

```python
support = dataset_results.support
```

```python
first = []
second = []
third = []
fourth = []
```

```python
for i in range(dataset_results.shape[0]):
    single_list = dataset_results['ordered_statistics'][i][0]
    first.append(list(single_list[0]))
    second.append(list(single_list[1]))
    third.append(single_list[2])
    fourth.append(single_list[3])
```

```python
lhs = pd.DataFrame(first)
rhs= pd.DataFrame(second)
confidence=pd.DataFrame(third,columns=['Confidence'])
lift=pd.DataFrame(fourth,columns=['lift'])
```

```python
datasetfinal = pd.concat([lhs,rhs,support,confidence,lift], axis=1)
datasetfinal
```

```python
datasetfinal.fillna(value=' ', inplace=True)
datasetfinal.columns = ['lhs',1,2,'rhs','support','confidence','lift']
datasetfinal['lhs'] = datasetfinal['lhs']+str(", ")+datasetfinal[1]+str(", ")+datasetfinal[2]
datasetfinal.drop(columns=[1,2],inplace=True)
datasetfinal
```
